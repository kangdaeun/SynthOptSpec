```python
myteff = 4500.
def get_spec_file(teff):
    teffstr = str(int(teff/100))
    if teff<1000.:
        teffstr = '00'+teffstr
    elif teff<10000.:
        teffstr = '0'+teffstr
    Z = '0'
    LogG = '5'
    specfile = 'bt-settl/lte'+teffstr+'-'+LogG+'-'+Z+'a+0.0.BT-NextGen.7.dat.txt'
    #specfile = 'testdata/test.dat.txt'
    return specfile

#R=4000.

specfile = get_spec_file(myteff)

testpar = {
    'file' : specfile,
    'wlmin' : 4000.,
    'wlmax' : 9800.,
    'wledge' : 10.,
    'R' : 4000.,
    'modflux_log' : False,
    'format' : 'txt',
}
#print(testpar['file'])
testpar['file'] = '../../testdata/lte035-3.0-0.0a+0.0.BT-NextGen.7.dat.txt'

test_fits = testpar.copy()
test_fits['format'] = 'fits'
test_fits['file'] = "../../testdata/lte035.0-3.0-0.0a+0.0.BT-Settl.spec.fits"

test_atlas9 = testpar.copy()
test_atlas9['format'] = 'atlas9'
test_atlas9['file'] = '../../testdata/fp00t3500g30k2odfnew.fl.dat.txt'
```

```python

# define spectrum reading the file
test_ng_035_30 = SynSpec(testpar)
test_st_035_30  = SynSpec(test_fits)
#test_a9_035_30 = SynSpec(test_atlas9)
# smooth to a resolution R

test_ng_035_30.smoothspec(testpar['R'])
test_st_035_30.smoothspec(testpar['R'])
# resample
wlres = np.arange(4000.,9800.,1.3)
test_ng_035_30.resample(wlres,smoothed=True)
test_st_035_30.resample(wlres,smoothed=True)

# plot the result
test_ng_035_30.plotspec(smoothed=True, resampled=True, showedge=True)
#test_st_035_30.plotspec(smoothed=True, resampled=True, showedge=True)

plt.plot(test_st_035_30.rswl,test_st_035_30.rsfl/10.,color='g')
```
